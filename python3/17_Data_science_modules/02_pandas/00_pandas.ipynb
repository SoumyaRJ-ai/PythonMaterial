creating a dataframe from a dict

```python
try:
    my_dict = {
        "Name": ["Tom", "nick", "krish", "jack"],
        "Age": [20, 21, 19, 18, 89],
    }  # has one value extra

    df = pd.DataFrame(my_dict)
except ValueError as ex:
    print(ex)
```

```python
my_dict = {
    "Name": ["udhay", "Tom", "nick", "krish", "jack"],
    "Age": [20, 21, 19, 18, 89],
}  # has one value extra

df = pd.DataFrame(my_dict)
df
```

Working with missing data

```python
dict = {
    "First Score": [100, 90, np.nan, 95],
    "Second Score": [30, 45, 56, np.nan],
    "Third Score": [np.nan, 40, 80, 98],
}

df_dict = pd.DataFrame(dict)
df_dict
```

```python
df_dict.isnull()
```

Filling missing values using fillna(), replace() and interpolate()

```python
df_dict.fillna(0)
```

```python
df_dict  # original df is not changed
```

```python
df_dict.dropna()
```

```python
df_dict  # orginal df is not changed
```

Iterating over the dataframe

```python
for each in df_dict:
    print(each)
```

```python
for each in df_dict.items():
    print(each, "\n")
```

```python
for key, value in df_dict.items():
    print(key, "===>", value, "\n")  # complete column data is value
```
