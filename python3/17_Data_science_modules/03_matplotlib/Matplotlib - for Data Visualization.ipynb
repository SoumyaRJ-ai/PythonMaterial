```python
# deferred rendering mechanism

import numpy as np
import matplotlib.pyplot as plt


def plotSlope(X, Y):
    Xs = X[1:] - X[:-1]
    Ys = Y[1:] - Y[:-1]
    plt.plot(X[1:], Ys / Xs)


X = np.linspace(-3, 3, 100)
Y = np.exp(-(X**2))

plt.plot(X, Y)
plotSlope(X, Y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = [[5.0, 25.0, 50.0, 20.0], [4.0, 23.0, 51.0, 17.0], [6.0, 22.0, 52.0, 19.0]]

color_list = ["b", "g", "r"]
gap = 0.8 / len(data)

for i, row in enumerate(data):
    X = np.arange(len(row))
    plt.bar(X + i * gap, row, width=gap, color=color_list[i % len(color_list)])

plt.show()
```

### PLotting Stacked bar charts

```python
import matplotlib.pyplot as plt

A = [5.0, 30.0, 45.0, 22.0]
B = [5.0, 25.0, 50.0, 20.0]

X = range(4)

plt.bar(X, A, color="b")
plt.bar(X, B, color="g", bottom=A)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([5.0, 30.0, 45.0, 22.0])
B = np.array([5.0, 25.0, 50.0, 20.0])
C = np.array([1.0, 2.0, 1.0, 1.0])
X = np.arange(4)

plt.bar(X, A, color="b")
plt.bar(X, B, color="g", bottom=A)
plt.bar(X, C, color="r", bottom=A + B)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array(
    [[5.0, 30.0, 45.0, 22.0], [5.0, 25.0, 50.0, 20.0], [1.0, 2.0, 1.0, 1.0]]
)

color_list = ["b", "g", "r"]
X = np.arange(data.shape[1])

for i in range(data.shape[0]):
    plt.bar(
        X,
        data[i],
        bottom=np.sum(data[:i], axis=0),
        color=color_list[i % len(color_list)],
    )

plt.show()
```

### Back-to-back bar charts

### Plotting Triangulation
* Triangulations arise when dealing with spatial locations. Apart from showing distances
between points and neighborhood relationships, triangulation plots can be a convenient
way to represent maps. matplotlib provides a fair amount of support for triangulations.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.tri as tri

data = np.random.rand(100, 2)
triangles = tri.Triangulation(data[:, 0], data[:, 1])
plt.triplot(triangles)
plt.show()
```

### Defining own colors for plots

```python
import numpy as np
import matplotlib.pyplot as plt


def pdf(X, mu, sigma):
    a = 1.0 / (sigma * np.sqrt(2.0 * np.pi))
    b = -1.0 / (2.0 * sigma**2)
    return a * np.exp(b * (X - mu) ** 2)


X = np.linspace(-6, 6, 1000)

for i in range(5):
    samples = np.random.standard_normal(50)
    mu, sigma = np.mean(samples), np.std(samples)
    plt.plot(X, pdf(X, mu, sigma), color=".75")

plt.plot(X, pdf(X, 0.0, 1.0), color="k")
plt.show()
```

### Using custom colors for scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.random.standard_normal((100, 2))
A += np.array((-1, -1))  # Center the distrib. at <-1, -1>

B = np.random.standard_normal((100, 2))
B += np.array((1, 1))  # Center the distrib. at <1, 1>

plt.scatter(A[:, 0], A[:, 1], color=".25")
plt.scatter(B[:, 0], B[:, 1], color=".75")
plt.show()
```

### Using colormaps for scatter plots
* colormaps define colors with a continous function of one variable to one value, corresponding to one color.

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

N = 256
angle = np.linspace(0, 8 * 2 * np.pi, N)
radius = np.linspace(0.5, 1.0, N)

X = radius * np.cos(angle)
Y = radius * np.sin(angle)

plt.scatter(X, Y, c=angle, cmap=cm.hsv)
plt.show()
```

### Using colormap for bar charts

```python
import numpy as np
import matplotlib.cm as cm
import matplotlib.colors as col
import matplotlib.pyplot as plt

values = np.random.randint(low=0, high=99, size=50)
cmap = cm.ScalarMappable(col.Normalize(0, 99), cm.binary)

plt.bar(np.arange(len(values)), values, color=cmap.to_rgba(values))
plt.show()
```

### Controlling line pattern and thickness

```python
import numpy as np
import matplotlib.pyplot as plt


def pdf(X, mu, sigma):
    a = 1.0 / (sigma * np.sqrt(2.0 * np.pi))
    b = -1.0 / (2.0 * sigma**2)
    return a * np.exp(b * (X - mu) ** 2)


X = np.linspace(-6, 6, 1024)
plt.plot(X, pdf(X, 0.0, 1.0), color="k", linestyle="solid")
plt.plot(X, pdf(X, 0.0, 0.5), color="k", linestyle="dashed")
plt.plot(X, pdf(X, 0.0, 0.25), color="k", linestyle="dashdot")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 8
A = np.random.random(N)
B = np.random.random(N)
X = np.arange(N)

plt.bar(X, A, color=".75")
plt.bar(X, A + B, bottom=A, color="w", linestyle="dashed")
plt.show()
```
